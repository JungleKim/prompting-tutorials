# 퓨샷 프롬프팅 (Few-shot Prompting)

퓨샷 프롬프팅은 모델에게 몇 가지 예시를 제공하여 특정 패턴이나 형식을 학습시킨 후 새로운 입력에 대해 유사한 결과를 생성하도록 하는 기법입니다. 이 방식은 특히 특수한 형식이나 스타일이 필요한 작업에 유용합니다.

이 노트북에서는 퓨샷 프롬프팅의 기본 개념을 배우고 다양한 작업에 적용해보겠습니다.

````python
# 필요한 라이브러리 및 모듈 임포트
import os
import sys
from dotenv import load_dotenv

# utils 디렉토리의 helpers.py 모듈을 임포트하기 위한 경로 설정
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from utils.helpers import get_completion, create_few_shot_prompt

# .env 파일 로드
load_dotenv()
````

## 1. 기본 퓨샷 프롬프팅

퓨샷 프롬프팅의 기본 구조는 다음과 같습니다:
1. 지시사항(지시문)
2. 예시 입력과 출력 쌍 여러 개
3. 새로운 입력

이러한 구조를 통해 모델은 지시사항을 구체적인 예시와 함께 이해하고, 새로운 입력에 대해 일관된 형식과 스타일로 응답할 수 있습니다.

````python
# 기본 퓨샷 프롬프팅 예시: 감정 분석
prompt = """
다음은 영화 리뷰의 감정을 분석하는 예시입니다. 리뷰의 감정을 '긍정', '부정', '중립' 중 하나로 분류해주세요.

리뷰: "이 영화는 정말 재미있었어요. 배우들의 연기도 훌륭했고 스토리도 흥미로웠습니다."
감정: 긍정

리뷰: "시간 낭비였습니다. 스토리도 없고 배우들의 연기도 별로였어요."
감정: 부정

리뷰: "기대했던 것보다는 덜 재미있었지만, 몇몇 장면은 인상적이었습니다."
감정: 중립

리뷰: "처음에는 지루했지만 후반부로 갈수록 흥미진진해졌어요. 결말이 약간 아쉽습니다."
감정:
"""

response = get_completion(prompt)
print(response)
````

## 2. 포맷 지정 퓨샷 프롬프팅

````python
# 포맷 지정 퓨샷 프롬프팅 예시: JSON 형식 변환
prompt = """
다음은 사람에 대한 설명을 JSON 형식으로 변환하는 예시입니다. 주어진 설명을 이름, 나이, 직업, 취미를 포함한 JSON 객체로 변환해주세요.

설명: "김민수는 32세 소프트웨어 엔지니어이며 등산과 독서를 좋아합니다."
JSON: {"이름": "김민수", "나이": 32, "직업": "소프트웨어 엔지니어", "취미": ["등산", "독서"]}

설명: "박지연은 28세 의사이고 요리와 여행을 즐깁니다."
JSON: {"이름": "박지연", "나이": 28, "직업": "의사", "취미": ["요리", "여행"]}

설명: "이준호는 45세 교사이며 사진 촬영과 음악 감상이 취미입니다."
JSON:
"""

response = get_completion(prompt)
print(response)
````

````python
# 다단계 추론 퓨샷 프롬프팅 예시: 수학 문제 해결
prompt = """
다음은 수학 문제를 단계별로 해결하는 예시입니다. 각 문제를 단계별로 풀어서 최종 답을 구해주세요.

문제: "상자 안에 사과 8개와 배 12개가 있습니다. 영희가 과일의 3분의 1을 먹었다면 남은 과일은 몇 개인가요?"

단계 1: 상자 안의 총 과일 수를 구합니다.
사과 8개 + 배 12개 = 20개

단계 2: 영희가 먹은 과일의 수를 구합니다.
총 과일 수 20개의 3분의 1 = 20 ÷ 3 = 6.67개 → 반올림하면 7개

단계 3: 남은 과일의 수를 구합니다.
총 과일 수 20개 - 영희가 먹은 과일 7개 = 13개

답: 13개

문제: "한 학급에 학생이 30명 있습니다. 남학생은 여학생보다 4명 더 많습니다. 이 학급의 여학생은 몇 명인가요?"

단계 1:
"""

response = get_completion(prompt)
print(response)
````

## 7. 퓨샷 프롬프팅 최적화 팁

효과적인 퓨샷 프롬프팅을 위한 몇 가지 팁을 정리해보았습니다:

1. **대표적인 예시 선택**: 다양하고 대표적인 예시를 선택하여 모델이 패턴을 더 잘 파악할 수 있게 합니다.
2. **예시 개수 최적화**: 너무 많은 예시는 불필요하게 토큰을 소모하고, 너무 적은 예시는 효과가 떨어질 수 있습니다. 작업 복잡도에 따라 2-5개 사이의 예시가 적절합니다.
3. **형식 일관성 유지**: 모든 예시에서 동일한 입력-출력 형식을 유지하여 모델이 패턴을 쉽게 학습하도록 합니다.
4. **난이도 순서 배열**: 쉬운 예시부터 복잡한 예시 순으로 배열하면 모델의 이해도를 점진적으로 높일 수 있습니다.
5. **경계 조건 포함**: 일반적인 케이스뿐만 아니라 예외적인 경우나 경계 조건도 포함시켜 모델의 대응력을 향상시킵니다.
6. **작업 분해**: 복잡한 작업은 여러 단계로 분해하여 각 단계별로 예시를 제공합니다.
7. **명확한 지시어 사용**: 예시와 함께 명확한 지시어를 포함하여 모델이 과제를 정확히 이해할 수 있게 합니다.
8. **도메인 지식 활용**: 전문 분야의 작업은 해당 분야의 용어와 맥락을 포함한 예시를 제공합니다.

이러한 팁을 활용하여 퓨샷 프롬프팅의 효과를 극대화할 수 있습니다.
