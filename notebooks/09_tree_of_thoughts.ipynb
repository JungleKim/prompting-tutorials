# Tree of Thoughts: 대규모 언어 모델을 위한 체계적 문제 해결 기법

Tree of Thoughts(ToT)는 Chain of Thought 프롬프팅을 일반화한 프레임워크로, 언어 모델이 더 체계적이고 신중한 추론을 할 수 있도록 돕습니다.

## 주요 특징

- **탐색적 추론**: 여러 가능한 사고 경로를 동시에 탐색
- **자기 평가**: 각 중간 단계의 품질을 스스로 평가
- **전략적 계획**: 앞을 내다보고 필요시 백트래킹
- **전역 최적화**: 지역적 선택이 아닌 전체적인 관점에서 의사결정

## 성능

Game of 24와 같은 수학 퍼즐에서 GPT-4의 성공률을 4%에서 74%로 크게 향상시켰습니다.
