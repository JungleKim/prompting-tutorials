# 제로샷 프롬프팅 (Zero-shot Prompting)

제로샷 프롬프팅은 모델에게 어떠한 예시도 제공하지 않고 직접 작업을 수행하도록 지시하는 기법입니다. 이 방식은 모델이 기본적으로 갖추고 있는 지식과 능력을 테스트하는 데 유용합니다.

이 노트북에서는 제로샷 프롬프팅의 기본 개념을 배우고 다양한 작업에 적용해보겠습니다.

```python
# 필요한 라이브러리 및 모듈 임포트
import os
import sys
from dotenv import load_dotenv

# utils 디렉토리의 helpers.py 모듈을 임포트하기 위한 경로 설정
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from utils.helpers import get_completion

# .env 파일 로드
load_dotenv()

print(os.getenv("MODEL"))
```

## 1. 기본 제로샷 프롬프팅

가장 간단한 형태의 제로샷 프롬프팅은 모델에게 직접적으로 지시를 내리는 것입니다. 예를 들어, 텍스트 요약, 감정 분석, 질문 답변 등의 작업을 수행하도록 요청할 수 있습니다.

```python
# 기본 제로샷 프롬프팅 예시: 텍스트 요약
text = """
인공지능(AI)은 인간의 학습, 추론, 지각, 언어 이해 등의 지능을 컴퓨터 프로그램으로 구현한 기술입니다.
딥러닝, 머신러닝, 자연어 처리 등 다양한 기술이 AI의 발전을 이끌고 있으며,
최근에는 ChatGPT, DALL-E, MidJourney 등의 생성형 AI 모델이 큰 주목을 받고 있습니다.
AI는 의료, 금융, 교육, 교통 등 다양한 분야에 활용되고 있으며 사회와 경제에 큰 변화를 가져오고 있습니다.
그러나 AI의 발전에 따라 윤리적 문제, 일자리 대체, 개인정보 보호 등의 과제도 함께 대두되고 있습니다.
"""

prompt = f"다음 텍스트를 한 문장으로 요약해주세요: {text}"

response = get_completion(prompt)
print(response)
```

```python
# 기본 제로샷 프롬프팅 예시: 감정 분류
prompt = """
Text를 중립, 부정 또는 긍정으로 분류합니다.
Text: 휴가는 괜찮다고 생각합니다.
Sentiment:
"""

### 감정 (중립으로 도출)
response = get_completion(prompt)
print(response)
```

## 2. 작업 형식 지정하기

제로샷 프롬프팅을 할 때 출력 형식을 지정하면 더 구조화된 응답을 얻을 수 있습니다.

```python
# 출력 형식을 지정한 제로샷 프롬프팅
prompt = f"""
다음 텍스트의 감정을 분석하고, 결과를 다음 형식으로 출력해주세요:
- 감정: [긍정/부정/중립]
- 확신도: [0-10 사이의 점수]
- 주요 감정 단어: [텍스트에서 감정을 나타내는 주요 단어 3개]
- 간단한 이유: [한 문장으로 분석 이유 설명]

텍스트: "오늘 시험 결과를 받았는데 예상보다 점수가 잘 나와서 정말 기뻐요! 열심히 공부한 보람이 있네요."
"""

response = get_completion(prompt)
print(response)
```

## 3. 역할 부여하기

모델에게 특정 역할을 부여하면 그 역할에 맞는 전문적인 응답을 얻을 수 있습니다.

```python
# 역할 부여를 활용한 제로샷 프롬프팅
prompt = """
당신은 프로그래밍 교육 전문가입니다. 초보자가 파이썬을 처음 배우려고 합니다.
파이썬에서 변수, 자료형, 조건문, 반복문의 기본 개념을 설명하고,
간단한 예제 코드와 함께 초보자를 위한 학습 로드맵을 제시해주세요.
"""

response = get_completion(prompt)
print(response)
```

## 4. 복잡한 작업 분해하기

복잡한 작업은 단계별로 분해하여 지시하면 더 좋은 결과를 얻을 수 있습니다.

```python
# 복잡한 작업을 단계별로 분해한 제로샷 프롬프팅
prompt = """
다음 텍스트를 분석하여 주요 논점을 파악하고, 비판적으로 평가해주세요. 다음 단계별로 진행해주세요:

1. 먼저 텍스트의 주요 주장과 근거를 파악하세요.
2. 주장의 논리적 타당성을 평가하세요.
3. 주장에 사용된 증거의 신뢰성과 충분성을 평가하세요.
4. 반대 의견이나 대안적 관점을 제시하세요.
5. 전체적인 논지의 강점과 약점을 요약하세요.

텍스트: "인공지능의 발전은 장기적으로 많은 일자리를 대체할 것이다. 역사적으로 기술 발전은 항상 특정 직업군을 없애왔지만, 동시에 새로운 직업을 창출해왔다. 그러나 인공지능은 이전의 기술혁명과 달리 인간의 인지 능력을 대체할 수 있기 때문에, 모든 산업 분야에서 광범위한 자동화가 일어날 것이다. 따라서 우리는 보편적 기본소득과 같은 새로운 사회 시스템을 고려해야 한다."
"""

response = get_completion(prompt)
print(response)
```

## 5. 제로샷 프롬프팅의 한계

제로샷 프롬프팅은 간단하고 일반적인 작업에는 효과적이지만, 특수한 형식이나 복잡한 추론이 필요한 작업에서는 한계가 있습니다. 이런 경우에는 few-shot 프롬프팅이나 chain-of-thought 프롬프팅과 같은 더 고급 기법이 필요할 수 있습니다.

다음은 제로샷 프롬프팅의 한계를 보여주는 예시입니다.

```python
# 제로샷 프롬프팅의 한계 예시
prompt = """
다음 수학 문제를 풀어주세요:

1, 3, 6, 10, 15, ... 이 수열의 100번째 항은 무엇인가요?
"""

response = get_completion(prompt)
print(response)

# 이 예시에서는 모델이 수열의 패턴을 정확히 파악하지 못하거나 계산 과정에서 오류를 범할 수 있습니다.
# 이런 경우 few-shot이나 chain-of-thought 방식이 더 효과적일 수 있습니다.
```

## 6. 실습: 다양한 작업에 제로샷 프롬프팅 적용하기

이제 여러분이 직접 다양한 작업에 제로샷 프롬프팅을 적용해보세요. 아래 셀을 활용하여 다음 작업들을 시도해보세요:

1. 텍스트 분류 (뉴스 기사의 카테고리 분류)
2. 창의적인 글쓰기 (특정 주제나 스타일로 짧은 이야기 작성)
3. 코드 생성 (간단한 프로그래밍 문제 해결)
4. 질문 답변 (특정 주제에 대한 질문에 답변)

각 작업마다 효과적인 프롬프트를 만들어보고, 결과를 분석해보세요.

```python
# 여기에 여러분의 제로샷 프롬프팅 실험을 작성하세요
# 예: 텍스트 분류
news_article = """
삼성전자가 새로운 플래그십 스마트폰 '갤럭시 S25'를 내년 초 출시할 계획이라고 밝혔다.
이번 신제품은 강화된 AI 기능과 접이식 디스플레이 기술을 탑재해 시장의 큰 관심을 모으고 있다.
업계 관계자들은 이번 출시가 글로벌 스마트폰 시장의 경쟁 구도에 변화를 가져올 것으로 전망했다.
"""

prompt = f"""
다음 뉴스 기사를 가장 적절한 카테고리로 분류해주세요:
- 정치
- 경제
- 사회
- 문화
- 스포츠
- IT/과학
- 국제

기사: {news_article}

카테고리: 
"""

response = get_completion(prompt)
print(response)
```

## 7. 제로샷 프롬프팅 최적화 팁

효과적인 제로샷 프롬프팅을 위한 몇 가지 팁을 정리해보았습니다:

1. **명확한 지시**: 모델에게 무엇을 원하는지 명확하게 지시합니다.
2. **출력 형식 지정**: 원하는 출력 형식을 구체적으로 명시합니다.
3. **역할 부여**: 특정 전문가나 역할을 부여하여 응답의 전문성을 높입니다.
4. **단계적 접근**: 복잡한 작업은 단계별로 분해하여 지시합니다.
5. **한계 인식**: 제로샷 프롬프팅의 한계를 인식하고, 필요시 다른 기법을 활용합니다.

이러한 팁을 활용하여 제로샷 프롬프팅의 효과를 극대화할 수 있습니다.
